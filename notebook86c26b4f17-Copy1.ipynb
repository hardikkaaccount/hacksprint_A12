```python
import re

def debias_text(input_text, biased_term_mapping):
    
    pattern = re.compile(r'\b(?:%s)\b' % '|'.join(map(re.escape, biased_term_mapping.keys())), re.IGNORECASE)

    debiased_text = pattern.sub(lambda match: biased_term_mapping[match.group(0).lower()], input_text)

    return debiased_text
```

```python
def debias_list_of_strings(list_of_strings, biased_term_mapping):
    debiased_list = [debias_text(text, biased_term_mapping) for text in list_of_strings]
    return debiased_list
```

```python
biased_term_mapping = {
    'he': 'they',
    'him': 'them',
    'his': 'their',
    'man': 'person',
    'woman': 'individual',
    'women': 'people',
    'black': 'person of color',
    'white': 'person',
    'old': 'experienced',
    'young': 'early in their career',
    'liberal': 'open-minded',
    'conservative': 'cautious',
    'left-wing': 'progressive',
    'right-wing': 'traditional',
    'poor': 'financially disadvantaged',
    'rich': 'financially privileged',
    'lower class': 'economically disadvantaged',
    'upper class': 'economically privileged',
    'disabled': 'differently abled',
    'normal': 'typical',
    'handicap': 'challenged',
    'gay' : 'same-sex attracted',
    'lesbian': 'same-sex attracted woman',
    'transgender': 'gender diverse',
    'queer': 'non-heteronormative',
    'fat': 'body positive',
    'skinny': 'slender',
    'obese': 'having a larger body size',
    'crazy': 'emotionally distressed',
    'normal': 'typical',
    'psycho': 'person with a mental health condition'
    # Add more biased terms and their debiased counterparts
}
```

```python
def recommend(news):
    index = new[new['title'] == news].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])

    data = []
    for i in distances[0:8]:
        item = []
        # print(new.iloc[i[0]].title)
        temp_df = newses[newses['Unnamed: 0.1'] == new.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Unnamed: 0.1')['title'].values))
        # item.extend(list(temp_df.drop_duplicates('Unnamed: 0.1')['title_summary'].values))
        item.extend(list(temp_df.drop_duplicates('Unnamed: 0.1')['date'].values)) 
        item.extend(list(temp_df.drop_duplicates('Unnamed: 0.1')['link'].values)) 

        data.append(item)
    return data
```
